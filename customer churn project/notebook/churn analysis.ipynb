## creating a summary table

```python
summary={}
summary["Total Customers"]=len(df_clean)
summary["Churn rate"]=(df_clean["Churn"].value_counts(normalize=True)["Yes"]*100)
summary["Average monthly charge"]=(df_clean["MonthlyCharges"].mean())
summary["Average Tenure"]=(df_clean["tenure"].mean())
summary
```

## Create New Business Features

### customer type

```python
df_clean["customerType"]=(df_clean["tenure"].apply(lambda x: "New" if x < 12 else "Loyal"))
```

```python
df_clean.head()
```

```python
df_clean["customerType"].value_counts()
```

```python
pd.crosstab(
df_clean["customerType"],
df_clean["Churn"],
normalize="index"
)*100
```

new customers churn more


## revenue segmentation

```python
df_clean["RevenueGroup"]=pd.cut(df_clean["MonthlyCharges"],bins=[0,35,70,120],labels=["Low","Medium","High"])
```

```python
df_clean[["MonthlyCharges","RevenueGroup"]].head()
```

### Analyze Churn by Segment

```python
pd.crosstab(
df_clean[
"RevenueGroup"
],
df_clean[
"Churn"
],
normalize="index"
)*100
```

High-paying customers leave more.

### Contract Analysis 

```python
df_clean.groupby("Contract")["MonthlyCharges"].agg(["count","mean"])
```

month to month contract have more customers

month to month contract earns most

Insight 1:
Month-to-month customers show highest churn.

Insight 2:
Customers with short tenure leave more.

Insight 3:
Higher monthly charges correlate with churn.
